```python
from langchain_community.utilities.dalle_image_generator import DallEAPIWrapper
from langchain.tools import Tool
from PIL import Image

# Initialize the DALL-E API Wrapper
dalle_api = DallEAPIWrapper(api_key=open_ai_key)

# Function to generate images using DALL-E
def generate_image(prompt):
    try:
        # Generate the image
        response = dalle_api.run(prompt)

        # Print the full response for debugging

        if response:
            return response
        else:
            return "Image URL not found in response"
    except Exception as e:
        return f"An error occurred: {e}"

# Define the tool for CrewAI
dalle_tool = Tool(
    name="dalle-image-generator",
    func=generate_image,  # Use the wrapper function
    description="Useful for generating images from text",
    result_as_answer=True,
)

# Example usage of the tool
if __name__ == '__main__':
    prompt = "Water in Moon"
    print(dalle_tool.run(prompt))  # Test the tool
```
